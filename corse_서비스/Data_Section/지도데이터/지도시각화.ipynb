```python
# # # ======= main main function ========# # # ==========================# # # ==========================# # #



def makemap(inp):
    gu_input = inp
    # 선택된 구 데이터만
    coord_gu = coord[coord['시군구']==gu_input]
    # 구 중심
    centerloc_n = centerloc['center_n'][gu_input]
    centerloc_e = centerloc['center_e'][gu_input]
    # map 생성
    maps = folium.Map(location=[centerloc_n, centerloc_e], zoom_start=13)

    # point - group 설명
    feature_group0 = folium.FeatureGroup(str(gu_input)+" 상권 중심 위치",show=False)
    for grp_name, df_grp in coord_gu.groupby('시군구_상권구분코드명'):
        # groupby > groupby 대상 , 나눠진 나머지 데이터프레임
        for point in coord_gu.index:
                icon_color="blue"
                mlocation = [coord_gu.loc[point, 'n'], coord_gu.loc[point, 'e']]
                # iframe 사용하여 popup width 고정
                iframe = folium.IFrame(coord_gu.loc[point, '시군구_상권코드명'])
                mpopup = folium.Popup(iframe,
                         min_width=200,
                         max_width=200)

                folium.CircleMarker(location=mlocation,
                                    popup=mpopup,
                                    color='blue',
                                    fill_color='black').add_to(feature_group0)
    feature_group0.add_to(maps)
    


    # 구-분기 코로나 확진자 시각화

    # legend delete
    def del_legend(choropleth):
        for child in choropleth._children:
            if child.startswith('color_map'):
                choropleth._children.pop(child)

    choropleth1 = folium.Choropleth(
        geo_data=state_geo,
        name='2020년 1분기 코로나 확진자 분포',
        data=corona_num['2020-1'],
        columns=[corona_num.index, corona_num['2020-1']],
        key_on= 'feature.properties.SIG_KOR_NM',
        fill_color='OrRd',
        fill_opacity=0.7,
        line_color='purple',
        line_opacity=0.2,
        legend_name='2020년 1분기 코로나 환자 수 (명)',
        highlight = True,
        show=False
    ).add_to(maps)
    del_legend(choropleth1)
    choropleth1.geojson.add_child(
        folium.features.GeoJsonTooltip(["2020-1"], labels=False)
    )

    choropleth2 = folium.Choropleth(
        geo_data=state_geo,
        name='2020년 2분기 코로나 확진자 분포',
        data=corona_num['2020-2'],
        columns=[corona_num.index, corona_num['2020-2']],
        key_on= 'feature.properties.SIG_KOR_NM',
        fill_color='OrRd',
        fill_opacity=0.7,
        line_color='purple',
        line_opacity=0.2,
        legend_name='2020년 2분기 코로나 환자 수 (명)',
        highlight = True,
        show=False
    ).add_to(maps)
    del_legend(choropleth2)
    choropleth2.geojson.add_child(
        folium.features.GeoJsonTooltip(["2020-2"], labels=False)
    )

    choropleth3 = folium.Choropleth(
        geo_data=state_geo,
        name='2020년 3분기 코로나 확진자 분포',
        data=corona_num['2020-3'],
        columns=[corona_num.index, corona_num['2020-3']],
        key_on= 'feature.properties.SIG_KOR_NM',
        fill_color='OrRd',
        fill_opacity=0.7,
        line_color='purple',
        line_opacity=0.2,
        legend_name='2020년 3분기 코로나 환자 수 (명)',
        highlight = True,
        show=False
    ).add_to(maps)
    del_legend(choropleth3)
    choropleth3.geojson.add_child(
        folium.features.GeoJsonTooltip(["2020-3"], labels=False)
    )

    choropleth4 = folium.Choropleth(
        geo_data=state_geo,
        name='2020년 4분기 코로나 확진자 분포',
        data=corona_num['2020-4'],
        columns=[corona_num.index, corona_num['2020-4']],
        key_on= 'feature.properties.SIG_KOR_NM',
        fill_color='OrRd',
        fill_opacity=0.7,
        line_color='purples',
        line_opacity=0.2,
        legend_name='2020년 4분기 코로나 환자 수 (명)',
        highlight = True,
        show=False
    ).add_to(maps)
    del_legend(choropleth4)
    choropleth4.geojson.add_child(
        folium.features.GeoJsonTooltip(["2020-4"], labels=False)
    )

    choropleth5 = folium.Choropleth(
        geo_data=state_geo,
        name='2021년 1분기 코로나 확진자 분포',
        data=corona_num['2021-1'],
        columns=[corona_num.index, corona_num['2021-1']],
        key_on= 'feature.properties.SIG_KOR_NM',
        fill_color='OrRd',
        fill_opacity=0.7,
        line_color='purple',
        line_opacity=0.2,
        legend_name='2021년 1분기 코로나 환자 수 (명)',
        highlight = True,
        show=False
    ).add_to(maps)
    del_legend(choropleth5)
    choropleth5.geojson.add_child(
        folium.features.GeoJsonTooltip(["2021-1"], labels=False)
    )

    choropleth6 = folium.Choropleth(
        geo_data=state_geo,
        name='2021년 2분기 코로나 확진자 분포',
        data=corona_num['2021-2'],
        columns=[corona_num.index, corona_num['2021-2']],
        key_on= 'feature.properties.SIG_KOR_NM',
        fill_color='OrRd',
        fill_opacity=0.7,
        line_color='purple',
        line_opacity=0.2,
        legend_name='2021년 2분기 코로나 환자 수 (명)',
        highlight = True,
        show=False
    ).add_to(maps)
    del_legend(choropleth6)
    choropleth6.geojson.add_child(
        folium.features.GeoJsonTooltip(["2021-2"], labels=False)
    )

    choropleth7 = folium.Choropleth(
        geo_data=state_geo,
        name='2021년 3분기 코로나 확진자 분포',
        data=corona_num['2021-3'],
        columns=[corona_num.index, corona_num['2021-3']],
        key_on= 'feature.properties.SIG_KOR_NM',
        fill_color='OrRd',
        fill_opacity=0.7,
        line_color='purple',
        line_opacity=0.2,
        legend_name='2021년 3분기 코로나 환자 수 (명)',
        highlight = True,
        show=False
    ).add_to(maps)
    del_legend(choropleth7)
    choropleth7.geojson.add_child(
        folium.features.GeoJsonTooltip(["2021-3"], labels=False)
    )


    choropleth9 = folium.Choropleth(
        geo_data=state_geo,
        name='2021년 3분기까지 총 코로나 확진자 분포',
        data=corona_num['2020-2021 총합'],
        columns=[corona_num.index, corona_num['2020-2021 총합']],
        key_on= 'feature.properties.SIG_KOR_NM',
        fill_color='OrRd',
        fill_opacity=0.7,
        line_color='purple',
        line_opacity=0.2,
        legend_name='2020-2021 총합 코로나 환자 수 (명)',
        highlight = True,
        show=True
    ).add_to(maps)

    del_legend(choropleth9)

    # GeoJsonTooltip - mouseover
    choropleth9.geojson.add_child(
        folium.features.GeoJsonTooltip(["2020-2021 총합"], labels=False)
    )


    # 행정 구역 구분
    folium.GeoJson(state_geo,
                   name="서울 행정구역 구분",
                   zoom_on_click=True, 
                   show=False,
                   ).add_to(maps)
    
    
    # layercontrol
    folium.LayerControl(collapsed=False).add_to(maps)

    # 맵 저장
    filename="maphtml/"+gu_input+".html"
    maps.save(filename)

    print(gu_input)
    maps  
# 맵 구별 저장
for i in ["강남구","강동구","강북구","강서구","관악구","광진구","구로구","금천구","노원구","도봉구",
          "동대문구","동작구","마포구","서대문구","서초구","성동구","성북구","송파구","양천구","영등포구",
          "용산구","은평구","종로구","중구","중랑구"]:
    makemap(i)
print("end")

# makemap("중구")
# 구별 geojson 설정 -noneed
# 그룹 layer control bug 수정 -end
# 아이콘 변경 -end
# --구map.html 로 저장 -end
# iframe 으로 연결 -end
# autoscale 추가
# Choropleth mouseover설정 -end
# Choropleth legend 해제 -end
# corona legend 제거 -end
```
